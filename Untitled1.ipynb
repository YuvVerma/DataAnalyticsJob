```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset
df = pd.read_csv('/content/DataAnalyst.csv')
```

```python
#Inspect the data
df.head(5)
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df["Salary Estimate"] = df["Salary Estimate"].astype(str)
df["Salary Estimate"] = df["Salary Estimate"].str.replace('K', '', regex=True)
df["Salary Estimate"] = df["Salary Estimate"].str.extract(r'(\$\d+-\$\d+)')[0]
df[["Min Salary", "Max Salary"]] = df["Salary Estimate"].str.extract(r'\$(\d+)-\$(\d+)').astype(float) * 1000
df["Max Salary"] = df["Max Salary"].fillna(df["Min Salary"])
df["Avg Salary"] = (df["Min Salary"]+df["Max Salary"])/2
df[["Salary Estimate","Min Salary","Max Salary","Avg Salary"]].head()
```

```python
df.describe()
```

```python
print("\n Value Count of'Job title':\n",df['Job Title'].value_counts(dropna=False).head(10))
print("\n Value Count of 'Job Location':\n",df['Location'].value_counts(dropna=False).head(10))
```

```python
# Job location distribution
plt.figure(figsize=(10,5))
sns.countplot(y=df["Location"], order=df["Location"].value_counts().index[:10],palette="Set2")
plt.title("Top 10 Cities for Data Analyst Jobs")
plt.show()
```

```python
df_sort = df.sort_values(by="Avg Salary",ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x="Avg Salary",y="Job Title",data=df_sort,orient='h',order=df_sort["Job Title"].value_counts().head(10).index, palette="Set2")
plt.xlabel("Average Salary $")
plt.ylabel("Job Titel")
plt.title("Average salary by Job Title")
plt.show()
```

```python
plt.figure(figsize=(10,20))
sns.boxplot(y=df["Industry"],x=df["Avg Salary"])
plt.xticks(rotation=90)
plt.title("Salary Distribution Across Industries")
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_companies = df["Company Name"].value_counts().nlargest(10)
sns.barplot(x=top_companies.values, y=top_companies.index, palette="viridis")
plt.xlabel("Number of Job Listings")
plt.ylabel("Company Name")
plt.title("Top 10 Companies Hiring the Most Data Analysts")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Avg Salary'], kde=True)
plt.title('Distribution of Average Salary')
plt.xlabel('Average Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.boxplot(x=df["Size"],y=df["Avg Salary"], palette="coolwarm")
plt.title("Salary trends across different company sizes")
plt.xlabel("Company Size")
plt.ylabel("Average Salary (USD)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Rating"],bins=20, kde=True)
plt.title("Distribution of company ratings")
plt.xlabel("company rating")
plt.ylabel("Number of Comapnies")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["Rating","Min Salary","Max Salary","Avg Salary"]].corr(), annot=True, cmap="coolwarm", linewidths=1)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
top_jobs =df['Job Title'].value_counts().head(10)
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette="Set2")
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.title('Top 10 Jobs')
plt.show()
```

**Salary and Job Location:** Average salaries vary significantly by state, with California having the highest average and Utah having the lowest among the displayed states.

**Salary and Seniority:** Senior-level positions have considerably higher average salaries than junior or unspecified seniority levels.

**Salary and Company Age:** There's a very weak positive correlation (approximately 0.0015) between average salary and company age, suggesting minimal impact.

**Top Job Titles:**"Data Analyst" has the highest average salary among the top 10 most frequent job titles.
